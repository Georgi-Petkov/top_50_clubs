```python
# Load libraries

import pandas as pd
import numpy as np
import re
import requests 
from bs4 import BeautifulSoup 
import plotly.express as px
```

```python
# assign the url to scrape

URL = "https://www.forbes.com/sites/mikeozanian/2021/05/07/worlds-most-valuable-sports-teams-2021/?sh=1af1ab253e9e"
r = requests.get(URL) 

soup = BeautifulSoup(r.content, 'html5lib') 
#print(soup.prettify())
```

```python
data = []
for s in soup.find_all("strong"):
    data.append(''.join(s.get_text()))
```

```python
df = pd.DataFrame(np.array_split(data[2:-2], 50),
            columns=['club','value','five_year_change_in_value','owner','year_purchased','price_paid'])
```
